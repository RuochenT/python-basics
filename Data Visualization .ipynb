## 1. Simple Line Graph

```python
import matplotlib.pyplot as plt
squares = [1,4,9,16,25]
fig,ax = plt.subplots()
ax.plot(squares)

plt.show()
```

- squares to hold the data that we will plot. 
- the subplots() function generate one or more plots in the same figure.
- fig represents the entire figure (set properties that affect the entire figure, such as size, overall title, or layout.)
- ax represents a single plot in the figure (control properties related to a specific plot, such as axis labels, titles, data plotting, and ticks.)
- plot() method tries to plot the data it's given in a meaningful way.
- plt.show() opens Matplotlib's viewer and displays the plot.

### Changing the label type and line thickness

```python
fig, ax = plt.subplots()
ax.plot(squares, linewidth=3)

ax.set_title("Square Numbers",fontsize=24)
ax.set_xlabel("Value",fontsize=14)
ax.set_ylabel("Square of Value",fontsize=14)

ax.tick_params(labelsize=14)

plt.show()
```

- the linewidth parameter controls the thickness of the line that plot() generates
- set_title()method sets an overall title for the chart.
- the fontsize parameters control the size of the text in various elements in the chart.
- tick_params() stypes the tick mark labels to 14 on both axes.

## 2. Random Walks 

### Creating the RandomWalk Class
To make a random decisions, we store possible moves in a list and use the choice() function (from the random module) to decide which move to make each time a step is taken. 

```python
from random import choice 

class RandomWalk:
    """A class to generate random walks."""
    
    def __init__(self, num_points =5000):
        """Initialize attributes of a walk."""
        self.num_points = num_points
        
        # All walks start at (0,0).
        self.x_values = [0]
        self.y_values = [0]
    
    def fill_walk(self):
        """Calculate all the points in the walk."""
        # keep taking steps until the walk reaches the desired length. 
        while len(self.x_values) < self.num_points:
            
            x_direction = choice([1,-1])
            x_distance = choice([0,1,2,3,4])
            x_step = x_direction * x_distance 
            
            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4])
            y_step = y_direction * y_distance
            
            # Reject moves that go nowhere.
            if x_step == 0 and y_step ==0:
                continue 
                
            # calculate the new position based on the previous position.
            x = self.x_values[-1] + x_step 
            y = self.y_values[-1] + y_step
            
            # store the values by appending to x_values,y_values
            
            self.x_values.append(x)
            self.y_values.append(y)
```

### Plotting the Random walk

```python
import matplotlib.pyplot as plt

rw = RandomWalk()
rw.fill_walk()

# Plot the points in the walk.
plt.style.use('classic')
fig, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values, s=10)
ax.set_aspect('equal')
plt.show()
```

### Generating Multiple Random Walks

```python
while True:
    rw = RandomWalk()
    rw.fill_walk()
    
    plt.style.use('classic')
    fig, ax = plt.subplots()
    ax.scatter(rw.x_values, rw.y_values, s=10)
    ax.set_aspect('equal')
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

### Styling the Walk 

```python
rw = RandomWalk()
rw.fill_walk()
plt.style.use('classic')
fig, ax = plt.subplots()
point_numbers = range(rw.num_points)
ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues,
           edgecolors='none',s=15)
ax.set_aspect('equal')
plt.show()
```

We use range() to generate a list of numbers equal to the number of points in the walk 1. We assign this list to point_numbers, which well use to set the color of each point in the walk

**15-5: Refactoring**

```python
from random import choice 

class RandomWalk:
    """A class to generate random walks."""
    
    def __init__(self, num_points =5000):
        """Initialize attributes of a walk."""
        self.num_points = num_points
        
        # All walks start at (0,0).
        self.x_values = [0]
        self.y_values = [0]
    
    def fill_walk(self):
        """Calculate all the points in the walk."""
        # keep taking steps until the walk reaches the desired length. 
        while len(self.x_values) < self.num_points:
            
            x_step = self.get_step()
            y_step = self.get_step()
            
            # Reject moves that go nowhere.
            if x_step == 0 and y_step ==0:
                continue 
                
            # calculate the new position based on the previous position.
            x = self.x_values[-1] + x_step 
            y = self.y_values[-1] + y_step
            
            # store the values by appending to x_values,y_values
            
            self.x_values.append(x)
            self.y_values.append(y)
    
    def get_step(self):
        """calculate a single step in the walk"""
        direction = choice([1,-1])
        distance = choice([0,1,2,3,4])
        step = direction * distance
        return step
```

```python
import matplotlib.pyplot as plt

rw = RandomWalk(5000)
rw.fill_walk()
fig, ax = plt.subplots()
ax.plot(rw.x_values, rw.y_values, lw = 2, c=(0.1,0.6,1))

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.show()
```

## 3. Interactive plot with Plotly Express 
https://plotly.com/python/plotly-express

```python
from random import randint 

class Die:
    """A class representing a single die."""
    
    def __init__(self, num_sides=6):
        """Assume a six-side die."""
        self.num_sides = num_sides
    
    def roll(self):
        """Return a random value between 1 and number of sides"""
        return randint(1,self.num_sides)
```

We create an instance of Die with the default six sides. Then, we roll the die 100 times and store the result of each roll in the list 'results'.

```python
die = Die()
# Make some rolls, and store results in a list.
results = []
for roll_num in range(100): 
    result = die.roll()
    results.append(result)
    
print(results)
```

```python
die = Die()
results = []
for roll_num in range(1000):
    result = die.roll()
    results.append(result)
    
# Analyze the results 
frequencies = []
poss_results = range(1,die.num_sides+1)
for value in poss_results:
    frequency = results.count(value) # count value (1-6)
    frequencies.append(frequency)

print(frequencies)
```

### Making a Histogram 

```python
import plotly.express as px

title = "Results of Rolling One D6 1,000 Times"
labels = {'x':'Result', 'y': 'Frequency of Result'}
fig = px.bar(x=poss_results,y=frequencies, title=title,labels=labels)
fig.show()
```

### Rolling Two Dice

```python
import plotly.express as px

die_1 = Die()
die_2 = Die()

results = []
for roll_num in range(1000):
    result = die_1.roll() + die_2.roll()
    results.append(result)
    
# Analyze the results.
frequencies = []
max_result = die_1.num_sides + die_2.num_sides
poss_results = range(2,max_result+1)
for value in poss_results:
    frequency = results.count(value) # count 2 to 12
    frequencies.append(frequency)
    
# Visualize the results 
title = "Results of Rolling Two D6 Dice 1,000 Times"
labels = {'x':'Result','y':'Frequency of Result'}
fig = px.bar(x = poss_results, y = frequencies, title = title, labels = labels)

fig.update_layout(xaxis_dtick=1) # gives every bar its own label.

fig.show()

```
